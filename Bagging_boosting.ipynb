```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
le=LabelEncoder()
df['species']=le.fit_transform(df['species'])
```

```python
df['species'].value_counts()
```

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100,
                                  max_depth= None,
                                  random_state=42)
```

```python
rf_model.fit(X_train,y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model_ada = AdaBoostClassifier(n_estimators=100, random_state=42)
```

```python
model_ada.fit(X_train,y_train)
```

```python
y_pred_ada = model_ada.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_ada)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_gd = GradientBoostingClassifier(n_estimators=100,random_state=42)
```

```python
model_gd.fit(X_train,y_train)
```

```python
y_pred_gd = model_gd.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_gd)
```

```python
from xgboost import XGBClassifier
```

```python
model_xg = XGBClassifier(n_estimators=100,random_state=42,learning_rate=0.1,max_depth=3,use_label_encoder=False,eval_metric='mlogloss',)
```

```python
model_xg.fit(X_train,y_train)
```

```python
y_pred = model_xg.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
