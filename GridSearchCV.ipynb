```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
X = df.drop('species',axis=1)
y=df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=9)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
model_svm = SVC(C=30,gamma='auto',kernel='linear')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Classifier = GridSearchCV(model_svm,param_grid={
    'C':[1,10,20,30],
    'kernel':['linear','rbf'],
},cv = 5,return_train_score=False)
```

```python
Classifier.fit(X,y)
```

```python
Classifier.cv_results_
```

```python
df_class1 = pd.DataFrame(Classifier.cv_results_)
df_class1
```

```python
df_class1[['param_C','param_kernel','mean_test_score']]
```

```python
Classifier2 = GridSearchCV(model_knn,param_grid={
    'n_neighbors':[3,5,7,9],
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan''minkowski']
},cv=5,return_train_score=False)
```

```python
Classifier2.fit(X,y)
```

```python
df_class2 = pd.DataFrame(Classifier2.cv_results_)
df_class2
```

```python
df_class2.columns
```

```python
df_class2[['param_metric','param_n_neighbors','param_weights','mean_test_score']].sort_values(by='mean_test_score',ascending=False).head(1)
```
