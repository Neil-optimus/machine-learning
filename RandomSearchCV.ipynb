```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
X = df.drop('species',axis=1)
y=df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=9)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
model_svm = SVC(C=30,gamma='auto',kernel='linear')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rclassifier1 = RandomizedSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['linear','rbf'],
    'gamma':['auto','scale']
},cv=5,n_iter=10,return_train_score=False)
```

```python
rclassifier1.fit(X,y)
```

```python
rclassifier1.cv_results_
```

```python
r_class1_df = pd.DataFrame(rclassifier1.cv_results_)
```

```python
r_class1_df
```

```python
r_class1_df.columns
```

```python
r_class1_df[['param_kernel','param_gamma','param_C','mean_test_score']].sort_values(by='mean_test_score',ascending=False).head(1)
```

```python
rClassifier2 = RandomizedSearchCV((model_knn),{
    'n_neighbors':[1,3,5,7,9],
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan','minkowski']
},cv=5,n_iter=10,return_train_score=False)
```

```python
rClassifier2.fit(X,y)
```

```python
r_class2 = rClassifier2.cv_results_
```

```python
r_class2_df = pd.DataFrame(r_class2)
```

```python
r_class2_df
```

```python
r_class2_df.columns
```

```python
r_class2_df[['param_weights','param_n_neighbors','param_metric','mean_test_score']].sort_values(by='mean_test_score',ascending=False).head(1)
```
